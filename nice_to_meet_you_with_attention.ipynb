## 2. 데이터 전처리 (One-Hot Encoding)

````python
# One-Hot Encoding 함수
def one_hot_encode(indices, vocab_size):
    one_hot = np.zeros((len(indices), vocab_size), dtype=np.float32)
    for i, idx in enumerate(indices):
        one_hot[i, idx] = 1.0
    return one_hot
````

````python
# 입력과 출력 데이터 생성
input_sequences = []

for i, _ in index_to_char.items():
    input_sequences.append(i)
````

````python
class TransDataset(Dataset):
    def __init__(self, x,y):
        self.x0 = x[0]
        self.x1 = x[1]
        self.y  = y

    def __len__(self):
        return 1

    def __getitem__(self, index):
        return (self.x0, self.x1), self.y
        
````

````python
dataset = TransDataset(x,y)
dataloader = DataLoader(dataset, batch_size=1, shuffle=True)
print('the number of images', len(dataset))  
#print("----batch size",options.batchSize)
````

````python
class Encoder(nn.Module):
    def __init__(self, input_dim=4, hid_dim=6):
        super().__init__()
        self.hid_dim = hid_dim
        self.rnn = nn.LSTM(input_dim, hid_dim)
        
    def forward(self, src):
        
        # embedded = [src len, batch size, emb dim]
        outputs, (hidden, cell) = self.rnn(src)
        
        # outputs = [src len, batch size, hid dim * n directions]
        # hidden = [n layers * n directions, batch size, hid dim]
        # cell = [n layers * n directions, batch size, hid dim]
        
        return outputs, (hidden, cell)
````

````python
class Decoder(nn.Module):
    def __init__(self, output_dim=6, hid_dim=6):
        super().__init__()
        
        self.output_dim = output_dim
        self.hid_dim = hid_dim
        
        self.rnn = nn.LSTM(output_dim, hid_dim)
        
    def forward(self, src, hidden, cell):
        
        
        output, (hidden, cell) = self.rnn(src, (hidden, cell))
        
        
        return output, hidden, cell
````

````python
class Attention(nn.Module):
    def __init__(self, hid_dim=12, output_dim=6):
        super().__init__()
        
        self.affine = nn.Linear(hid_dim, output_dim)
        
    def forward(self, enc_hs,dec_hs):
        """
        enc_hs, des_hs format is seq,batch,data_size
        """
        
        softmax = nn.Softmax(dim=1)

        #compute attention weight from dot-product
        alpha = torch.bmm(enc_hs.permute(1,0,2), dec_hs.permute(0,2,1))
        alpha = softmax(alpha)
        h2 = enc_hs.permute(1,0,2)*alpha
        new_context = torch.sum(h2,axis=1)
        dec_in = torch.cat([dec_hs[0], new_context],axis=1)  
        output = self.affine(dec_in).unsqueeze(0)

        #print(f"output shape {output.shape}")
        return output
    
````

````python
class Seq2Seq(nn.Module):
   def __init__(self, encoder, decoder):
       super().__init__()
       
       self.encoder = encoder
       self.decoder = decoder
       self.attention = Attention(decoder.hid_dim*2, decoder.hid_dim)
       #self.device = device
       #print(f"output_dim : {self.decoder.output_dim}")
       # Encoder와 Decoder의 hidden dim이 같아야 함
       assert encoder.hid_dim == decoder.hid_dim

       
   def forward(self, src, trg, teacher_force=False):
       # src = [src len, batch size]
       # trg = [trg len, batch size]
       
       trg_len = trg.shape[0]
       batch_size = trg.shape[1]
       trg_vocab_size = 6
       
       # decoder 결과를 저장할 텐서
       outputs = torch.zeros(trg_len, batch_size, trg_vocab_size)
       
       # Encoder의 마지막 은닉 상태가 Decoder의 초기 은닉상태로 쓰임
       enc_hs, (hidden, cell) = self.encoder(src)

       #get attention score
       
       # Decoder에 들어갈 첫 input은 <sos> 토큰
       input = trg[0, :]

       #print(f"input shape : {input.shape} - hidden.sahpe : {hidden.shape}")
       # target length만큼 반복
       # range(0,trg_len)이 아니라 range(1,trg_len)인 이유 : 0번째 trg는 항상 <sos>라서 그에 대한 output도 항상 0 
       softmax = nn.Softmax(dim=1)
       for t in range(0, trg_len):
           
           output, hidden, cell = self.decoder(input.unsqueeze(0), hidden, cell)
           #print(f"output.shape : {output.shape}", hidden.shape)
           # alpha = torch.bmm(enc_hs.permute(1,0,2),output.permute(0,2,1))
           # alpha = softmax(alpha)
           # h2 = enc_hs.permute(1,0,2)*alpha
            
           # new_context = torch.sum(h2,axis=1)
           # #print(f"enc_hs : {enc_hs.shape}, ouput : {output.shape}")

           # #print(f"{alpha} alpha.shape: {alpha.shape}, mv.shape : {nv.shape}")
           # dec_in = torch.cat([output[0], new_context],axis=1)  
           # output = self.affine(dec_in).unsqueeze(0)
           output = self.attention(enc_hs, output)
           
           outputs[t] = output

           output = softmax(output[0])
           top1 = torch.argmax(output,dim=1)
           out2 = torch.zeros(batch_size, trg_vocab_size)
           out2[0:batch_size, top1] = 1
           if teacher_force == False:
               outputs[t] = out2.unsqueeze(0)
           # techer_force = 1 = True이면 trg[t]를 아니면 top1을 input으로 사용
           if t < trg_len-1:
               input = trg[t+1] if teacher_force else out2
       
       return outputs
````

````python
encoder = Encoder()
decoder = Decoder(6,6)
model = Seq2Seq(encoder, decoder)
#model.cuda()#
model.train()
````
